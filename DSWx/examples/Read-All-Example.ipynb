```python
import geopandas as gpd
import pandas as pd
import boto3
```

```python
session = boto3.session.Session(profile_name='saml-pub')
s3 = session.resource('s3')
s3_client = session.client('s3')
```

```python
bucket_name = 'opera-validation-datasets'
root_folder_name = 'calval-database-dev/'
```

```python
obj = s3.Object(bucket_name,root_folder_name+'image_calc.geojson')
image_calc_data = obj.get()['Body']
image_calc = gpd.read_file(image_calc_data)
image_calc.head()
```

```python
obj = s3.Object(bucket_name,root_folder_name+'image.geojson')
image_data = obj.get()['Body']
image = gpd.read_file(image_data)
image.head()
```

```python
obj = s3.Object(bucket_name,root_folder_name+'site.geojson')
site_data = obj.get()['Body']
site = gpd.read_file(site_data)
site.head()
```

```python
db = image_calc.merge(image,on='image_name').merge(site,on='site_name')
db.head()
```
