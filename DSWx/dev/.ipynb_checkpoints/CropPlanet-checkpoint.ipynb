```python
import os
import geopandas as gpd
import rasterio
import rasterio.mask
from rasterio.plot import show
import boto3
```

```python
download_dir = './downloads/'
crop_dir = download_dir+'crop/'
```

```python
image_files = os.listdir(download_dir)
file = image_files[0]
file
```

```python
session = boto3.session.Session(profile_name='saml-pub')
s3 = session.resource('s3')
s3_client = session.client('s3')
```

```python
bucket_name = 'opera-calval-database-dswx'
```

```python
obj = s3.Object(bucket_name,'site.geojson')
site_table_data = obj.get()['Body']
sites = gpd.read_file(site_table_data)
sites.head()
```

```python
obj = s3.Object(bucket_name,'image.geojson')
image_table_data = obj.get()['Body']
images = gpd.read_file(image_table_data)
images.head()
```

```python
chip4_42_ID = images[images.site_name == '4_42'].image_name.iloc[[0]]
chip4_42_ID = chip4_42_ID.iloc[0]
print(chip4_42_ID)
```

```python
for file in image_files:
    if file[-3:] != 'tif':
        continue
    
    ID = file.split('_3B_Analytic')[0]
    if ID == chip4_42_ID:
        print(file)
        imagery = rasterio.open(download_dir+file)
        image_crs = imagery.crs
        chip = sites[sites.site_name == '4_42']
        chip['geometry'] = chip.geometry.to_crs(image_crs).buffer(500,join_style=2)
        
        with rasterio.open(download_dir+file) as src:
            out_image, out_transform = rasterio.mask.mask(src, chip.geometry, crop=True)
            out_meta = src.meta
            
        out_meta.update({"driver": "GTiff",
                 "height": out_image.shape[1],
                 "width": out_image.shape[2],
                 "transform": out_transform})

        with rasterio.open(crop_dir+file, "w", **out_meta) as dest:
            dest.write(out_image)
```
